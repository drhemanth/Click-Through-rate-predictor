# Importing the test data

```python
#importing the graphlab, csv libraries
import graphlab as gl
import csv

data_train = gl.SFrame.read_csv('Data folder/train.txt', delimiter='\t', header = False)
data_test = gl.SFrame.read_csv('Data folder/test.txt', delimiter='\t', header = False)

print ('Both data_train and data_test files imported')
```

# Naming the headers

```python

data_train.rename({'X1':'Click', 'X2':'Weekday', 'X3':'Hour', 'X4':'Timestamp', 'X5':'LogType',
                   'X6':'UserID', 'X7':'UserAgent', 'X8':'IP', 'X9':'Region', 'X10':'City',
                   'X11':'AdExchange', 'X12':'Domain', 'X13':'URL', 'X14':'A_URLID', 'X15':'AdSlotID',
                   'X16':'AdSlotW', 'X17':'AdSlotH', 'X18':'AdSlotVisible', 'X19':'AdSlotFormat', 'X20':'floorPrice',
                    'X21':'CreativeID', 'X22':'keyPageURL', 'X23':'AdvertiserID','X24':'UserTag'})
```

```python
data_test.rename({'X1':'Weekday', 'X2':'Hour', 'X3':'Timestamp', 'X4':'LogType',
                   'X5':'UserID', 'X6':'UserAgent', 'X7':'IP', 'X8':'Region', 'X9':'City',
                   'X10':'AdExchange', 'X11':'Domain', 'X12':'URL', 'X13':'A_URLID', 'X14':'AdSlotID',
                   'X15':'AdSlotW', 'X16':'AdSlotH', 'X17':'AdSlotVisible', 'X18':'AdSlotFormat', 'X19':'floorPrice',
                    'X20':'CreativeID', 'X21':'keyPageURL', 'X22':'AdvertiserID','X23':'UserTag'})
```

# Click percentage calculator and validating the data set

1. Over all CTR percentage

```python
print ('the overall click through rate is in percentage= ') 
print(data_train['Click'].mean() * 100 )
```

2. CTR for each variant for example UserID in this case

```python
data_train.groupby('UserID', {'CTR':gl.aggregate.MEAN('Click')})
```

3. Display variation (min, mean, max)  of each variable in data set  

```python
gl.canvas.set_target('ipynb')
data_train.show()
```

# Creating a logistic regression model

```python
#Model using all variables 
model = gl.logistic_classifier.create(data_train, target='Click', features=None,max_iterations=25)
```

```python
#predicting using test.txt
model.predict(data_test, output_type='probability').head(10) 
```
